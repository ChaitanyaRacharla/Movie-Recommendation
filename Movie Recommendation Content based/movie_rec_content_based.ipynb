```python
df = pd.read_csv('IMDB_Top250Indianmovies2_OMDB_Detailed.csv')
df.head()
```

```python
df['clean_plot'] = df['Plot'].str.lower()
df['clean_plot'] = df['clean_plot'].apply(lambda x: re.sub('[^A-Za-z]', ' ', x))
df['clean_plot'] = df['clean_plot'].apply(lambda x: re.sub('\s+', ' ', x))
df['clean_plot']
```

```python
df['clean_plot'] = df['clean_plot'].apply(lambda x: nltk.word_tokenize(x))
df['clean_plot']
```

```python
stop_words = nltk.corpus.stopwords.words('english')
plot = []
for sentence in df['clean_plot']:
    temp = []
    for word in sentence:
        if word not in stop_words and len(word) >= 3:
            temp.append(word)
    plot.append(temp)
df['clean_plot'] = plot
df['clean_plot']
```

```python
df.head()
```

```python
df['Genre'] = df['Genre'].apply(lambda x: x.split(','))
df['Director'] = df['Director'].apply(lambda x: x.split(','))
df['Actors'] = df['Actors'].apply(lambda x: x.split(',')[:4])
```

```python
df['Actors'][0]
```

```python
def clean(sentence):
    temp = []
    for word in sentence:
        temp.append(word.lower().replace(' ',''))
    return temp
```

```python
df['Genre'] = [clean(x) for x in df['Genre']]
df['Director'] = [clean(x) for x in df['Director']]
df['Actors'] = [clean(x) for x in df['Actors']]
```

```python
df['Actors'][0]
```

```python
columns = ['clean_plot', 'Genre', 'Director', 'Actors']
lst = []
for i in range(len(df)):
    temp = ''
    for col in columns:
        temp += ' '.join(df[col][i]) + ' '
    lst.append(temp.strip())
lst
```

```python
df['clean_input'] = lst
df = df[['Title','clean_input']]
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer().fit_transform(df['clean_input'])
tfidf
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim = cosine_similarity(tfidf, tfidf)
cos_sim
```

```python
index = pd.Series(df['Title'])
index.head()
```

```python
def recommend_movies(title):
    movies = []
    idx = index[index == title].index[0]
    scores = pd.Series(cos_sim[idx]).sort_values(ascending=False)
    top10 = list(scores.iloc[2:12].index)
    for i in top10:
        movies.append(df['Title'][i])
    print(movies)
```

```python
recommend_movies('M.S. Dhoni: The Untold Story')
```
